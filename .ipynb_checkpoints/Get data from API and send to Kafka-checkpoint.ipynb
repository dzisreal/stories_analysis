```python
from json import dumps
import requests
from time import sleep
from kafka import KafkaProducer
from kafka.errors import KafkaError
from datetime import datetime
from random import randint
import threading
```

# Get data from API and send to Kafka

```python
current_time = datetime.now()
print(type(current_time))
```

```python
# defining key/request url
key = "https://api.binance.com/api/v3/ticker/price"
producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                             value_serializer=lambda x: 
                             dumps(x).encode('utf-8'))
# send messages to kafka
def send_to_kafka():
    while True:
        data = requests.get(key)  
        lstCoins = data.json()
        current_time = datetime.now()
        for coin in lstCoins:
            coin["time"] = str(current_time)
            if float(coin["price"]) > 200:
                coin["volume"] = randint(100,100000)
            else:
                coin["volume"] = randint(10,1000)
            producer.send('topic_test', value=coin)
            data = coin["symbol"]
            time = coin["time"]
            print(coin)
        sleep(0.5)
```

```python
send_to_kafka()
```
